# labels

```python
x = np.array([10,42,1,77,23,34])
y = np.array([92,12,19,52,41,38])
```

```python
plt.plot(x, y)
plt.title("BAŞLIK")
plt.xlabel("X EKSENİ İSMİ")
plt.ylabel("Y EKSENİ İSMİ")
plt.grid() # IZGARA
```

# subplots

```python
x = np.array([10,42,1,77,23,34])
y = np.array([92,12,19,52,41,38])
plt.subplot(1, 2, 1)
plt.title("1")
plt.plot(x, y)
x = np.array(["8,8,9,9,10,15"])
y = np.array(["24,20,26,27,29,34"])
plt.subplot(1, 2, 2)
plt.title("2")
plt.plot(x, y)
```
